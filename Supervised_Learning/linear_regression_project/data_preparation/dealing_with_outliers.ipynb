```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scipy as sp
import seaborn as sns
```

```python
df = pd.read_csv('../data/Ames_Housing_Data.csv')
```

```python
df.select_dtypes(include=np.number).corr()['SalePrice'].sort_values()
```

```python
sns.scatterplot(x='Overall Qual', y='SalePrice', data=df)
```

```python
sns.scatterplot(x='Gr Liv Area', y='SalePrice', data=df)
```

These points are going to be droped:

```python
drop_ind = df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000)].index
```

```python
df = df.drop(drop_ind, axis=0)
```

```python
sns.scatterplot(x='Gr Liv Area', y='SalePrice', data=df)
```

```python
df.to_csv('../data/Ames_Housing_Data_without_outliers.csv')
```
